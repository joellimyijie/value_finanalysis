**1. Loading html**

The following balance sheets from yahoo finance (annual, expanded all in webpage) are extracted:
 [LULU](https://finance.yahoo.com/quote/LULU/balance-sheet?p=LULU) [NIKE](https://finance.yahoo.com/quote/NIKE/balance-sheet?p=NIKE) [GIII](https://finance.yahoo.com/quote/GIII/balance-sheet?p=GIII) [COLM](https://finance.yahoo.com/quote/COLM/financials?p=COLM) [ADS](https://finance.yahoo.com/quote/ADS.DE/balance-sheet?p=ADS.DE) [VFC](https://finance.yahoo.com/quote/VFC/financials?p=VFC) [UA](https://finance.yahoo.com/quote/UA/balance-sheet?p=UA) [SKX](https://finance.yahoo.com/quote/SKX/balance-sheet?p=SKX)
 
You may right click on the website and press `page source` to see the source codes. Conversely, you may press `Ctrl+S` to save as HTML. 

```python
# Read Adidas Balance Sheet from Yahoo Finance in the form of HTML using Beautifulsoup
# We will be extending the ETL for Adidas to other companies later

with open("ADS.html", encoding="utf8") as fp:
    soup_ADS = BeautifulSoup(fp, 'html.parser')

# Here's how it looks like
soup_ADS
```

**2. Extract and transform**

In the file, you will see that there is data table in `<div id="rrtable">`. You may read more [here](https://realpython.com/beautiful-soup-web-scraper-python/) to `find` Beautifulsoup by ID.

```python
# I only need the data table, which is between <div id="rrtable"> and </div>

soup_ADS_extract = soup_ADS.find(id="rrtable")
soup_ADS_extract
```

```python
# Transform into Pandas table
soup_ADS_df = pd.read_html(str(soup_ADS_extract))[0]
soup_ADS_df

# as you can see, it is still pretty ugly
```

```python
# also move last row to the top and promote as headers
soup_ADS_df = soup_ADS_df.apply(np.roll, shift=1)
soup_ADS_df.columns = soup_ADS_df.iloc[0]
soup_ADS_df
```

```python
# keep only the rows I need and reset index for cleanliness
# keep only latest data for analysis
soup_ADS_df_filtered = soup_ADS_df[(soup_ADS_df['Period Ending:']=='Total Current Assets') | (soup_ADS_df['Period Ending:']=='Total Current Liabilities') | (soup_ADS_df['Period Ending:']=='Total Long Term Debt') | (soup_ADS_df['Period Ending:']=='Retained Earnings (Accumulated Deficit)')]
soup_ADS_df_filtered = soup_ADS_df_filtered.reset_index(drop=True)
soup_ADS_df_filtered = soup_ADS_df_filtered.iloc[:, 0:2]
soup_ADS_df_filtered
```

```python
# unpivot it into a row
soup_ADS_df_transpose = soup_ADS_df_filtered.T
soup_ADS_df_transpose.columns = soup_ADS_df_transpose.iloc[0]
soup_ADS_df_transpose = soup_ADS_df_transpose.iloc[1:].reset_index(drop=True)
soup_ADS_df_transpose
```

```python
# putting the above steps together in a function so that we can apply to the original HTML file

def extract_fields(soup):
    soup_extract = soup.find(id="rrtable")
    soup_df = pd.read_html(str(soup_extract))[0]
    soup_df = soup_df.apply(np.roll, shift=1)
    soup_df.columns = soup_df.iloc[0]
    soup_df_filtered = soup_df[(soup_df['Period Ending:']=='Total Current Assets') | (soup_df['Period Ending:']=='Total Current Liabilities') | (soup_df['Period Ending:']=='Total Long Term Debt') | (soup_df['Period Ending:']=='Retained Earnings (Accumulated Deficit)')]
    soup_df_filtered = soup_df_filtered.reset_index(drop=True)
    soup_df_filtered = soup_df_filtered.iloc[:, 0:2]
    soup_df_transpose = soup_df_filtered.T
    soup_df_transpose.columns = soup_df_transpose.iloc[0]
    soup_df_transpose = soup_df_transpose.iloc[1:].reset_index(drop=True)
    return soup_df_transpose

extract_fields(soup_ADS)
```

**3. Combining and loading all company files into pd**

```python
# Repeat the above steps for all the HTML files into the same pandas table 
# (Ideally I will loop through url but I am working with offline files now)
# Add in stock name

with open("ADS.html", encoding="utf8") as fp:
    soup_ADS = BeautifulSoup(fp, 'html.parser')
ADS_df = extract_fields(soup_ADS)
ADS_df.insert(0, "Stock", ['ADS'], False)

with open("COLM.html", encoding="utf8") as fp:
    soup_COLM = BeautifulSoup(fp, 'html.parser')
COLM_df = extract_fields(soup_COLM)
COLM_df.insert(0, "Stock", ['COLM'], False)

with open("GIII.html", encoding="utf8") as fp:
    soup_GIII = BeautifulSoup(fp, 'html.parser')
GIII_df = extract_fields(soup_GIII)
GIII_df.insert(0, "Stock", ['GIII'], False)

with open("LULU.html", encoding="utf8") as fp:
    soup_LULU = BeautifulSoup(fp, 'html.parser')
LULU_df = extract_fields(soup_LULU)
LULU_df.insert(0, "Stock", ['LULU'], False)

with open("NIKE.html", encoding="utf8") as fp:
    soup_NIKE = BeautifulSoup(fp, 'html.parser')
NIKE_df = extract_fields(soup_NIKE)
NIKE_df.insert(0, "Stock", ['NIKE'], False)

with open("SKX.html", encoding="utf8") as fp:
    soup_SKX = BeautifulSoup(fp, 'html.parser')
SKX_df = extract_fields(soup_SKX)
SKX_df.insert(0, "Stock", ['SKX'], False)

with open("UA.html", encoding="utf8") as fp:
    soup_UA = BeautifulSoup(fp, 'html.parser')
UA_df = extract_fields(soup_UA)
UA_df.insert(0, "Stock", ['UA'], False)

with open("VFC.html", encoding="utf8") as fp:
    soup_VFC = BeautifulSoup(fp, 'html.parser')
VFC_df = extract_fields(soup_VFC)
VFC_df.insert(0, "Stock", ['VFC'], False)

stocks_df = pd.concat([ADS_df,COLM_df,GIII_df,LULU_df,NIKE_df,SKX_df,UA_df,VFC_df])
stocks_df
```

```python
# Replace all '-' with NaN

stocks_df = stocks_df.replace('-', np.nan)
stocks_df
```

**4. Applying Graham's approach**
Here are the principles for an Enterprising Investor:
1. Strong Financial condition:

         current assets at least 1.5 times current liabilities

         total debt to net current assets ratio less that 1.1

2. Earnings Stability

         positive earnings for at least 5 years

3. Currently pays a dividend

4. Current earnings greater than years ago

5. Stock price less than 120% of net tangible assets

I will apply principle 1 and assume if current year applies for principle 2. The rest requires information from the main page and other financial sheets. 

```python
# Principle 1a
def principle_1a(s):
    if pd.isnull(s['Total Current Assets']) | pd.isnull(s['Total Current Liabilities']):
        return False
    else:
        return (float(s['Total Current Assets']) / float(s['Total Current Liabilities']))> float(1.5)

stocks_df['Principle 1a']=stocks_df.apply(principle_1a,axis=1)
stocks_df
```

```python
# Principle 1b
def principle_1b(s):
    if pd.isnull(s['Total Current Assets']) | pd.isnull(s['Total Long Term Debt']):
        return False
    else:
        return (float(s['Total Long Term Debt']) / float(s['Total Current Assets'])) < float(1.1)

stocks_df['Principle 1b']=stocks_df.apply(principle_1b,axis=1)
stocks_df
```

```python
# Principle 2
def principle_2(s):
    if pd.isnull(s['Retained Earnings (Accumulated Deficit)']):
        return False
    else:
        return float(s['Retained Earnings (Accumulated Deficit)']) > int(0)

stocks_df['Principle 2']=stocks_df.apply(principle_2,axis=1)
stocks_df
```

**5. Visualisation**

```python
# 3D scatterplot for 3 principles

import matplotlib.pyplot as plt

stocks_df['x'] = np.where(stocks_df['Principle 1a'], 1, 0)
stocks_df['y'] = np.where(stocks_df['Principle 1b'], 1, 0)
stocks_df['z'] = np.where(stocks_df['Principle 2'], 1, 0)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points
ax.scatter(stocks_df['x'], stocks_df['y'], stocks_df['z'])

# Highlight points where all three values are True
true_points = stocks_df[(stocks_df['x'] == 1) & (stocks_df['y'] == 1) & (stocks_df['z'] == 1)]
ax.scatter(true_points['x'], true_points['y'], true_points['z'], color='green')

# Add labels
for i in range(len(stocks_df)):
    ax.text(stocks_df['x'].iloc[i], stocks_df['y'].iloc[i], stocks_df['z'].iloc[i], stocks_df['Stock'].iloc[i])

ax.set_xlabel('Principle 1a')
ax.set_ylabel('Principle 1b')
ax.set_zlabel('Principle 2')

plt.show()
```

As you can see from the chart above, **SKX and UA** are the only stocks that passed the principles 1a, 1b and 2. Hope my company's financial health can improve...
